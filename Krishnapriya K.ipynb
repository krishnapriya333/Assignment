# **5**.Check and handle outliers in at least 3 columns in the dataset


```python
import seaborn as sns
```

```python
sns.boxplot(x='Fare',data=df2)
```

```python
sns.boxplot(x='Age',data=df2)
```

```python
sns.boxplot(x='SibSp',data=df2)
```

```python
age = df2['Age']
Q1 = age.quantile(0.25)
Q3 = age.quantile(0.75)
IQR = Q3 - Q1
threshold = 1.5
lower_whisker = Q1 -(threshold*IQR)
upper_whisker = Q3 + (threshold*IQR)
df2['Age']=np.where(df2['Age']>upper_whisker,upper_whisker,np.where(df2['Age']<lower_whisker,lower_whisker,df2['Age']))
```

```python
sns.boxplot(x='Age',data=df2)
```

```python
Fare = df2['Fare']
Q1 = Fare.quantile(0.25)
Q3 = Fare.quantile(0.75)
IQR = Q3 - Q1
threshold = 1.5
lower_whisker = Q1 -(threshold*IQR)
upper_whisker = Q3 + (threshold*IQR)
df2['Fare']=np.where(df2['Fare']>upper_whisker,upper_whisker,np.where(df2['Fare']<lower_whisker,lower_whisker,df2['Fare']))
```

```python
sns.boxplot(x='Fare',data=df2)
```

```python
Sibsp = df2['SibSp']
Q1 = Sibsp.quantile(0.25)
Q3 = Sibsp.quantile(0.75)
IQR = Q3 - Q1
threshold = 1.5
lower_whisker = Q1 -(threshold*IQR)
upper_whisker = Q3 + (threshold*IQR)
df2['SibSp']=np.where(df2['SibSp']>upper_whisker,upper_whisker,np.where(df2['SibSp']<lower_whisker,lower_whisker,df2['SibSp']))
```

```python
sns.boxplot(x='SibSp',data=df2)
```
